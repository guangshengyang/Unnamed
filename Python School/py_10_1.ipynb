## 创建并访问Series对象

```python
# 1.1 创建如下表的Series数据对象，其中a-f为索引
s_dict = {"a": 30, "b": 25, "c": 27, "d": 41, "e": 25, "f": 34}
s_obj = pd.Series(s_dict)
s_obj
```

```python
# 1.2 增加数据27，索引为g
s_obj["g"] = 27
s_obj
```

```python
# 1.3 修改索引d对应的值为40
s_obj["d"] = 40
s_obj
```

```python
# 1.4 查询值大于27的数据
s_obj[s_obj > 27]
```

```python
# 1.5 删除位置为1-3的数据
s_obj.drop(s_obj.index[1 : 3 + 1])
```
