```python
import random

a = random.randint(0, 100)
print("A random integer have been generated by computer, now guess it!")
print(f'For debug, the number is {a}')


round = 1  # round number

while True:
    print(f'Round {round}: ')
    try:
        d = int(input("Guess it here："))
    except ValueError:
        print("It's a inter, in fact.")
    else:
        if a == d:
            print("Try again!")
            break
        elif a < d:
            print("No, not that big.")
        else:
            print("no, not that small.")
        round += 1

print("Congratulations! You're right!")
```
